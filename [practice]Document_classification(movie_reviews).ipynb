```python
#seperate the reviews to positive and negative

positive = [movie_reviews.words(fileid) for fileid in movie_reviews.fileids('pos')]

negative = [movie_reviews.words(fileid) for flieid in movie_reviews.fileids('neg')]

            
```

```python
print(positive[0][:10])
print(negative[0])
```

```python
def beg_of_words(words):
    cleaned = [w.lower() for w in words
              if w not in stopwords.words('english')
              if w not in string.punctuation]
    word_dic = dict([w, True] for w in cleaned)
    return word_dic
# make beg of words without stopwords and puctuation
beg_of_words(['I','am','a','a','groot'])
```

```python
# make two feature sets, positive review set, negative review set
pos_reviews_set = []
for words in positive:
    pos_reviews_set.append((beg_of_words(words), 'pos'))
 

neg_reviews_set = []
for words in negative:
    neg_reviews_set.append((beg_of_words(words), 'neg'))
 

print (pos_reviews_set[0])
```

```python
print(len(pos_reviews_set), len(neg_reviews_set))
```

```python
shuffle(pos_reviews_set)
shuffle(neg_reviews_set)
test_set = pos_reviews_set[:200] + neg_reviews_set[:200]
train_set = pos_reviews_set[200:] + neg_reviews_set[200:]
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_set)
```
